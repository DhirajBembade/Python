 # Accessing Elements of Tuple:
  We can access either by index or by slice operator


```python
#By using Index:
t=(10,20,30,40,50,60)
print(t[0])#10
print(t[-1])#60
print(t[100])#IndexError: tuple index out of range
```

```python
# By using Slice Operator:
t=(10,20,30,40,50,60)
print(t[2:5])
print(t[2:100])
print(t[::2])
```

#  Mathematical Operators for Tuple:
   We can apply + and * operators for tuple

```python
# 1) Concatenation Operator (+):
t1=(10,20,30)
t2=(40,50,60)
t3=t1+t2
print(t3) # (10,20,30,40,50,60)
```

```python
#2) Multiplication Operator OR Repetition Operator (*)

t1=(10,20,30)
t2=t1*3
print(t2)#(10,20,30,10,20,30,10,20,30)
```

# Important Functions of Tuple:

```python
#1. len():
t= (1,2,3,4)
len(t)
```

```python
#2. count():To return number of occurrences of given element in the tuple
t= (1,2,3,4,1,1)
t.count(1)
```

```python
# 3. index(): Returns index of first occurrence of the given element.
# If the specified element is not available then we will get ValueError.
t=(10,20,10,10,20)
t.index(20)
```

```python
# 4. sorted(): To sort elements based on default natural sorting order
t=(10,20,10,10,20)
t1 = sorted(t)
print(t1)
```

```python
#We can sort according to reverse of default natural sorting order as follows
t12 = sorted(t, reverse = True)
print(t12)
```

```python
# 5. min() and max() functions:
#These functions return min and max values according to default natural sorting order.
t=(10,20,30,10,10,20)
print(min(t))#10
print(max(t))#40
```

Q. What do you mean by packing and unpacking of tuple?(explain with the help of code)


* Packing tuples-

Packing tuples assigns multiple Python objects to a tuple.

tuple1 = (obj1, obj2, ..., objN)

Each object in tuples has an index and can be accessed using its index.

 Once an object has been assigned to a tuple, it cannot be changed, i.e., it is immutable.

* Unpacking tuples-

Unpacking tuples assigns the objects in a tuple to multiple variables.

(obj1, obj2, ..., objN) = tuple1
During unpacking, the number of variables must be equal to the number of objects in the tuples. If this is not the case, a ValueError is raised.


```python
var1 = 2
var2 = 0.4
var3 = "Hello"
var4 = [1, 2, 3, 4]

# Packing tuple
tuple1 = (var1, var2, var3, var4)
print("Tuple: ", tuple1)
```

```python
# Unpacking tuple
intObj, floatObj, strObj, listObj = tuple1
print("intObj: ", intObj)
print("floatObj: ", floatObj)
print("strObj: ", strObj)
print("listObj: ", listObj)
```
